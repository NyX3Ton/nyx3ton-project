```python
# Stretnutie 2, Domáca úloha

import math
import pandas as pd
from scipy import stats

df = pd.read_excel("./DUdata1a2.xlsx", sheet_name="HODNOTENIE SLUŽIEB")
x = df["hodnotenie"].dropna()

n = len(x)
x_bar = x.mean()
s = x.std(ddof=1)

mu0 = 4
alpha = 0.10

t_stat = (x_bar - mu0) / (s / math.sqrt(n))
t_crit = stats.t.ppf(1 - alpha, df=n-1)

print("Zistite priemerné hodnotenie služieb v danej vzorke =", round(x_bar, 3))
print("Testovacia štatistika pre vyhodnotenie hypotézy =", round(t_stat, 3))
print("Kvantil Studentovho rozdelenia pre vyhodnotenie hypotézy =", round(t_crit, 3))
print("Zamietam H0" if t_stat > t_crit else "Nezamietam H0")
```

```python
# Stretnutie 3, Domáca úloha

import math
import pandas as pd
from scipy.stats import f as f_dist
from scipy.stats import t as t_dist

FILE_PATH = "DomacaUloha3_.xlsx"
SHEET_NAME = "Žiarovky"
ALPHA = 0.10

df = pd.read_excel(FILE_PATH, sheet_name=SHEET_NAME)

df.columns = ["zivotnost", "metoda"]

stara = df[df["metoda"] == 0]["zivotnost"]
nova = df[df["metoda"] == 1]["zivotnost"]

n1 = len(stara)
n2 = len(nova)

mean1 = stara.mean()
mean2 = nova.mean()

var1 = stara.var(ddof=1)
var2 = nova.var(ddof=1)

print("1) Priemerna zivotnost starej metody:", round(mean1, 3))
print("2) Priemerna zivotnost novej metody:", round(mean2, 3))
print("3) Odhad rozptylu starej metody:", round(var1, 3))
print("4) Odhad rozptylu novej metody:", round(var2, 3))

if var1 >= var2:
    F = var1 / var2
    df_num = n1 - 1
    df_den = n2 - 1
else:
    F = var2 / var1
    df_num = n2 - 1
    df_den = n1 - 1

p_value_f = 2 * (1 - f_dist.cdf(F, df_num, df_den))

print("5) Testovacia statistika F:", round(F, 3))
print("   p-value pre F test:", round(p_value_f, 3))

if p_value_f < ALPHA:
    print("6) H0 o zhode rozptylov: zamietame")
    equal_var = False
else:
    print("6) H0 o zhode rozptylov: nezamietame")
    equal_var = True

if equal_var:
    sp2 = ((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2)
    t_stat = (mean1 - mean2) / math.sqrt(sp2 * (1/n1 + 1/n2))
    df_t = n1 + n2 - 2
else:
    t_stat = (mean1 - mean2) / math.sqrt(var1/n1 + var2/n2)
    df_t = ((var1/n1 + var2/n2) ** 2) / (
        ((var1/n1) ** 2) / (n1 - 1) + ((var2/n2) ** 2) / (n2 - 1)
    )

p_value_t = 2 * (1 - t_dist.cdf(abs(t_stat), df_t))

print("7) Testovacia statistika t:", round(t_stat, 3))
print("   p-value pre t-test:", round(p_value_t, 3))

if p_value_t < ALPHA:
    print("8) H0 o zhode priemernej zivotnosti: zamietame")
else:
    print("8) H0 o zhode priemernej zivotnosti: nezamietame")
```
