```python
###Hruby domaci produkt per capita
gdp_per_capita = eurostat.get_data_df('nama_10_pc')
gdp_per_capita = gdp_per_capita[(gdp_per_capita['geo\\TIME_PERIOD'].isin(['SK', 'CZ','EU27_2020','EA']) & (gdp_per_capita['na_item'] == 'B1GQ') & (gdp_per_capita['unit'] == 'CP_PPS_EU27_2020_HAB'))]
gdp_per_capita = gdp_per_capita[['unit','na_item', 'geo\\TIME_PERIOD','2007','2008','2009','2010','2011','2012','2013']]
gdp_per_capita['geo\\TIME_PERIOD'] = gdp_per_capita['geo\\TIME_PERIOD'].replace({'EA': 'Oblast Eura','EU27_2020': 'Staty EU'})
gdp_per_capita = gdp_per_capita.rename(columns={'geo\\TIME_PERIOD': 'Krajina', 'na_item': 'Indikator', 'unit' : 'Jednotka'})
gdp_per_capita['Indikator'] = gdp_per_capita['Indikator'].replace({'B1GQ': 'HDP podla trhovych cien'})
gdp_per_capita['Jednotka'] = gdp_per_capita['Jednotka'].replace({'CP_PPS_EU27_2020_HAB': 'PPS per capita'})

display(gdp_per_capita.tail(100))

gdp_1 = gdp_per_capita.melt(
    id_vars=['Krajina'],
    value_vars=['2007','2008','2009','2010','2011','2012','2013'],
    var_name='Year',
    value_name='GDP_per_capita')

fig = px.line(
    gdp_1,
    x='Year',
    y='GDP_per_capita',
    color='Krajina',
    markers=True,
    text='GDP_per_capita', 
    title='HDP per capita – SK vs CZ vs Oblast eura vs Staty EU (2007–2013)',
    labels={'GDP_per_capita': 'HDP per capita (PPS)', 'Year': 'Rok'}
)

fig.for_each_trace(lambda t: t.update(
    mode='lines+markers+text',
    line=dict(width=4),
    marker=dict(size=12, line=dict(width=2, color='white')),
    textposition='top center',
    texttemplate='%{text:.1f}'
))

fig.update_layout(template='simple_white', hovermode=False)
fig.show()


file_name = f'./Vykonnost ekonomiky/gdp_per_capita_{today}.csv'
gdp_per_capita.to_csv(os.getcwd() + r'.\Vykonnost ekonomiky\gdp_per_capita_{}.csv'.format(today),index=True, header=True, encoding="utf-8-sig")
```

```python
###Deficit/prebytok verejnych financií
statny_deficit = eurostat.get_data_df('gov_10dd_edpt1')
statny_deficit = statny_deficit[(statny_deficit['geo\\TIME_PERIOD'].isin(['SK', 'CZ','EU27_2020','EA20'])) & (statny_deficit['unit'] == 'PC_GDP') & (statny_deficit['sector'] == 'S13') & (statny_deficit['na_item'] == 'B9')]
statny_deficit = statny_deficit[['unit','sector','na_item', 'geo\\TIME_PERIOD','2007','2008','2009','2010','2011','2012','2013']]

statny_deficit['geo\\TIME_PERIOD'] = statny_deficit ['geo\\TIME_PERIOD'].replace({ 'EA20': 'Oblast Eura','EU27_2020': 'Staty EU'})
statny_deficit = statny_deficit.rename(columns={'geo\\TIME_PERIOD': 'Krajina', 'unit' : 'Jednotka', 'sector' : 'Statny sektor', 'na_item' : 'Polozka'})
statny_deficit['Jednotka'] = statny_deficit['Jednotka'].replace({'PC_GDP': '% z HDP'})
statny_deficit['Statny sektor'] = statny_deficit['Statny sektor'].replace({'S13': 'Vlada celkovo'})
statny_deficit['Polozka'] = statny_deficit['Polozka'].replace({'B9': 'Cisty dlh'})

file_name = f'./Fiskalna politika/statny_deficit_{today}.csv'
statny_deficit.to_csv(os.getcwd() + r'.\Fiskalna politika\statny_deficit_{}.csv'.format(today),index=True, header=True, encoding="utf-8-sig")
display(statny_deficit.tail(100))

statny_deficit = statny_deficit.melt(
    id_vars=['Krajina'],
    value_vars=['2007','2008','2009','2010','2011','2012','2013'],
    var_name='Rok',
    value_name='Statny_dlh')

fig = px.line(
    statny_deficit,
    x='Rok',
    y='Statny_dlh',
    color='Krajina',
    markers=True,
    text='Statny_dlh',
    title='Štátny deficit/prebytok – SK vs CZ (2007–2013)',
    labels={'Statny_dlh': 'Čistý dlh v % HDP', 'Rok': 'Rok'}
)

fig.for_each_trace(lambda t: t.update(
    mode='lines+markers+text',
    line=dict(width=4),
    marker=dict(size=12, line=dict(width=2, color='white')),
    textposition='top center',
    texttemplate='%{text:.1f}'
))

# Ensure all years show and in correct order
fig.update_xaxes(
    type='category',
    categoryorder='array',
    categoryarray=[str(y) for y in range(2007, 2014)]
)

# Horizontal Maastricht line across full width
fig.add_shape(
    type="line",
    xref="paper", yref="y",   # <<< key change
    x0=0, x1=1,               # span entire plotting area
    y0=-3, y1=-3,
    line=dict(width=3, color="red")
)
fig.add_annotation(
    x=0.01, y=-3, xref="paper", yref="y",
    text="Maastricht (−3 % HDP)",
    showarrow=False, yshift=-10,
    font=dict(size=12, color="red")
)

fig.update_layout(template='simple_white', hovermode=False)
fig.show()
```

```python
url = (
    "https://sdmx.oecd.org/public/rest/data/"
    "OECD.SDD.NAD,DSD_NAMAIN10@DF_TABLE1_EXPENDITURE,2.0/all"
    "?startPeriod=2007"
    "&endPeriod=2013"
    "&dimensionAtObservation=AllDimensions"
    "&format=csvfilewithlabels"
)

response = requests.get(url)
response.raise_for_status()

df = pd.read_csv(StringIO(response.text))

countries = ["SVK", "CZE"]
transactions = ["B1GQ", "P6", "P7"]   # HDP, export, import

df = df[
    (df["REF_AREA"].isin(countries)) &
    (df["TRANSACTION"].isin(transactions)) &
    (df["FREQ"] == "A")
].copy()

# Rok musí byť integer
df["TIME_PERIOD"] = df["TIME_PERIOD"].astype(int)

pivot_df = df.pivot_table(
    index=["REF_AREA", "TIME_PERIOD"],
    columns="TRANSACTION",
    values="OBS_VALUE"
).reset_index()

pivot_df = pivot_df.rename(columns={
    "REF_AREA": "country",
    "TIME_PERIOD": "year",
    "B1GQ": "GDP",
    "P6": "EXPORT",
    "P7": "IMPORT"
})

pivot_df["TRADE_BALANCE"] = pivot_df["EXPORT"] - pivot_df["IMPORT"]
pivot_df["OPENNESS_pct_GDP"] = (pivot_df["EXPORT"] + pivot_df["IMPORT"]) / pivot_df["GDP"] * 100

# Výsledný DataFrame
result_df = pivot_df.sort_values(["country", "year"]).reset_index(drop=True)

float_cols = ['EXPORT', 'IMPORT', 'TRADE_BALANCE', 'OPENNESS_pct_GDP']
result_df[float_cols] = result_df[float_cols].astype(np.float32).round(2)
result_df = result_df[['country','year','EXPORT', 'IMPORT', 'TRADE_BALANCE', 'OPENNESS_pct_GDP']]
file_name = f'./Medzinarodny obchod/import_export_{today}.csv'
result_df.to_csv(os.getcwd() + r'.\Medzinarodny obchod\import_export_{}.csv'.format(today),index=True, header=True, encoding="utf-8-sig")

trade_gph = result_df.copy()
trade_gph['year'] = trade_gph['year'].astype(str)  # peknejší x-axis

fig_trade = px.line(
    trade_gph,
    x='year',
    y='TRADE_BALANCE',
    color='country',
    markers=True,
    text='TRADE_BALANCE',
    title='Obchodná bilancia – CZE vs. SVK (2007–2013)',
    labels={
        'year': 'Rok',
        'TRADE_BALANCE': 'mil USD',
        'country': 'Krajina'
    }
)

fig_trade.for_each_trace(
    lambda tr: tr.update(
        mode='lines+markers+text',
        line=dict(width=4),
        marker=dict(size=12, line=dict(width=2, color='white')),
        textposition='top center',
        texttemplate='%{text:.0f}'  # celé čísla (sumy v mene)
    )
)

fig_trade.update_layout(
    template='simple_white',
    hovermode=False
)

open_gph = result_df.copy()
open_gph['year'] = open_gph['year'].astype(str)

fig_open = px.line(
    open_gph,
    x='year',
    y='OPENNESS_pct_GDP',
    color='country',
    markers=True,
    text='OPENNESS_pct_GDP',
    title='Miera otvorenosti ekonomiky – CZE vs. SVK (2007–2013)',
    labels={
        'year': 'Rok',
        'OPENNESS_pct_GDP': '% HDP',
        'country': 'Krajina'
    }
)

fig_open.for_each_trace(
    lambda tr: tr.update(
        mode='lines+markers+text',
        line=dict(width=4),
        marker=dict(size=12, line=dict(width=2, color='white')),
        textposition='top center',
        texttemplate='%{text:.1f}'  # 1 desatinné miesto pre %
    )
)

fig_open.update_layout(
    template='simple_white',
    hovermode=False
)


expimp_df = result_df.copy()
expimp_df['year'] = expimp_df['year'].astype(str)

# Melt → pripravíme 4 kombinácie: CZE/SVK × Export/Import
expimp_melt = expimp_df.melt(
    id_vars=['country', 'year'],
    value_vars=['EXPORT', 'IMPORT'],
    var_name='Type',
    value_name='Value'
)

# Pre krajšie texty
expimp_melt['Type'] = expimp_melt['Type'].replace({
    'EXPORT': 'Export',
    'IMPORT': 'Import'
})

# Spojíme do jednej legendy
expimp_melt['Legend'] = expimp_melt['country'] + " – " + expimp_melt['Type']

# Bar graf s klastrovaním vedľa seba
fig_bar = px.bar(
    expimp_melt,
    x='year',
    y='Value',
    color='Legend',
    barmode='group',
    text='Value',
    title='Export a import – CZE vs SVK (2007–2013)',
    labels={
        'year': 'Rok',
        'Value': 'mil USD',
        'Legend': 'Krajina – Tok'
    }
)

# Štýl podľa tvojej šablóny
fig_bar.for_each_trace(
    lambda tr: tr.update(
        texttemplate='%{text:.0f}',
        textposition='outside'
    )
)

fig_bar.update_layout(
    template='simple_white',
    hovermode=False,
    bargap=0.25,
    bargroupgap=0.05,
    uniformtext_minsize=8,
    uniformtext_mode='show'
)
fig_open.show()
fig_bar.show()
fig_trade.show()
display(result_df)
```
