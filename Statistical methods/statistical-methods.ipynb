```python
import math
import pandas as pd
from scipy import stats

df = pd.read_excel("./DUdata1a2.xlsx", sheet_name="HODNOTENIE SLUŽIEB")
x = df["hodnotenie"].dropna()

n = len(x)
x_bar = x.mean()
s = x.std(ddof=1)

mu0 = 4
alpha = 0.10

t_stat = (x_bar - mu0) / (s / math.sqrt(n))
t_crit = stats.t.ppf(1 - alpha, df=n-1)

print("priemer =", round(x_bar, 3))
print("t =", round(t_stat, 3))
print("t_kvantil =", round(t_crit, 3))
print("zamietam H0" if t_stat > t_crit else "nezamietam H0")
```
